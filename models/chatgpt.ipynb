```python
import requests
import tiktoken
import pandas as pd
import csv
import time

imgs_questions = pd.read_csv("../Dataset/SoMDataset/Relational/3_color_size_shape/3_color_size_shape.csv")
# change the path to your CSV file accordingly

# 修改下面的范围
imgurl_list = imgs_questions.iloc[:, 0]
text_list = imgs_questions.iloc[:, 1]


all_responses=[]

url =  ""

encoding = tiktoken.encoding_for_model("gpt-4o")

headers = {
    "Authorization": "Bearer YOUR_API_KEY",
    "Content-Type": "application/json",
}

for imgurl, text in zip(imgurl_list, text_list):
    data = {
        "model": "gpt-4o",
        "messages": [
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": text+",不使用序号，使用(0,0)到(10,10)内的坐标表示",
                    },
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": imgurl
                        },
                    },
                ],
            }
        ],
        "max_tokens": 300,
    }
    # 计算文本内容的token数
    text_content = data["messages"][0]["content"][0]["text"]
    num_tokens = len(encoding.encode(text_content))

    print(f"发送文本内容的token数: {num_tokens}")
    response = requests.post(url, headers=headers, json=data)
    if response.status_code == 200:
        content = response.json()['choices'][0]['message']['content']
        # 打印回复内容
        print(content)
    else:
        print(f"请求失败，状态码：{response.status_code}")



```
